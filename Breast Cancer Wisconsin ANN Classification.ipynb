```python
# Splitting The Dataset
```

```python
x = df.drop('benign_0__mal_1',axis=1).values
y = df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
# Scaling The data
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
```

```python
x_train.shape
```

```python
model = tf.keras.models.Sequential()

model.add(tf.keras.layers.Dense(units=30,activation='relu'))
model.add(tf.keras.layers.Dense(units=15,activation='relu'))
## Binary Classification thats why sigmoid activation function
model.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
model.fit(x_train,y_train,epochs=600,validation_data=(x_test,y_test))
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
# We observe, a case of overfitting to our training dataset
# Early Stopping by Callbacks
```

```python
# Redefining the model, so we dont train the old model again
model = tf.keras.models.Sequential()

model.add(tf.keras.layers.Dense(units=30,activation='relu'))
model.add(tf.keras.layers.Dense(units=15,activation='relu'))
## Binary Classification thats why sigmoid activation function
model.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# If metric we monitor was accuracy, we would want to maximize it so the mode = max.
# In this case, we are monitoring loss and we want to minimize it, mode=min
# Patience meaning we will wait 25 epochs even after it detected a stopping point becoz of noise that could have occured.
early_stop = EarlyStopping(monitor = 'val_loss', mode='min',verbose=1,patience=25)
```

```python
# Need to pass it as a list
model.fit(x_train,y_train,epochs=600,validation_data=(x_test,y_test),
          callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

```python
# This is okay
# We can do one more thing to prevent overfitting i.e dropoutlayers
```

```python
# Redefining the model, so we dont train the old model again
model = tf.keras.models.Sequential()

model.add(tf.keras.layers.Dense(units=30,activation='relu'))

# rate=0 means we are taking out nothing and turning no neurons off
# Very common value b/w 0.2 & 0.5
# 0.5 means half the neurons during each batch i.e 1 epoch training, half nuerons are going to be turned off randomly
model.add(tf.keras.layers.Dropout(rate=0.5))

model.add(tf.keras.layers.Dense(units=15,activation='relu'))

# here also
model.add(tf.keras.layers.Dropout(rate=0.5))

## Binary Classification thats why sigmoid activation function
model.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
model.fit(x_train,y_train,epochs=600,validation_data=(x_test,y_test),
          callbacks=[early_stop])
```

```python
# We can see it ran a little bit longer beacoz of the dropout layers
```

```python
model_loss = pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

```python
# Going down at same and flattening out also on the same rate
```

```python
## Prediction
```

```python
predictions = (model.predict(x_test) > 0.5)*1 
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```
